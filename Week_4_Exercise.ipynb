```python
#Exercise 2
def crop_grid(img, num_horizontal_grid, num_vertical_grid, line_color):
    # img is the source image
    img_copy = img.copy()
    
    # Convert line_color string to RGB value
    named_colors = {
        'black': (0, 0, 0),
        'white': (255, 255, 255),
        'red': (0, 0, 255),
        'green': (0, 255, 0),
        'blue': (255, 0, 0),
        # Add more named colors if needed
    }

    if line_color in named_colors:
        line_color = named_colors[line_color]
    else:
        raise ValueError("Unknown named color: " + line_color)
        
    # num_horizontal_grid and num_vertical_grid are the number of patches along x and y axes.
    height, width = img.shape[:2]
    
    M, N = int(height / num_vertical_grid), int(width / num_horizontal_grid)
    # line_color is the color of the grid line.
    x1, y1 = 0, 0
    for y in range(0, height, M):
         for x in range(0, width, N):

             y1 = y + M
             x1 = x + N

             if x1>=width and y1>=height:
                 x1 = width-1
                 y1 = height-1
                 tile = img[y:height, x:width]
                 cv.rectangle(img_copy, (x,y), (x1, y1), line_color, 1)
                 cv.imshow('tile', tile)

             elif y1>=height:
                 y1 = height-1
                 cv.rectangle(img_copy, (x, y), (x1, y1), line_color, 1)

             elif x1>=width:
                 x1 = width-1
                 cv.rectangle(img_copy, (x, y), (x1, y1), line_color, 1)

             else:
                 cv.rectangle(img_copy, (x, y), (x1, y1), line_color, 1)
    # The output of the function should be image with grids
    show_img('patched image', img_copy)
```

```python
#Exercise 3
import cv2

# Load the two base images
img1 = cv2.imread('images/lena.jfif')
img2 = cv2.imread('images/coins.jfif')

# Reshape img2
new_shape = img1.shape[:2]
img2 = cv.resize(img2, new_shape[::-1])

# Create a window to display the images
cv2.namedWindow('Image Transition', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Image Transition', 800, 400)

# Perform image blending with different alpha values
for alpha in range(0, 11):
    alpha = alpha / 10.0

    # Blend the images using alpha as the weight
    blended = cv2.addWeighted(img1, alpha, img2, 1 - alpha, 0)

    # Display the blended image
    cv2.imshow('Image Transition', blended)
    cv2.waitKey(200)  # Delay between frames (200 milliseconds)

cv2.destroyAllWindows()
```

```python
#Exercise 4
img = cv.imread("images/travel_hd.jpg")
watermark = cv.imread("images/watermark.jpeg")

#resize
img = cv.resize(img, None, fx = 0.1, fy = 0.1)
watermark = cv.resize(watermark, None, fx = 1.5, fy = 2.5)

#get background and watermark height and width
imgh, imgw = img.shape[:2]
h, w = watermark.shape[:2]

#create black background with the same size as background
blank_bg = np.zeros_like(img, np.uint8)
blank_bg[imgh-h:imgh, :w] = watermark #place watermark at bottom left
mask = blank_bg.astype(bool) #convert all black(zero) pixels to false

#blend only masks with true value
img[mask] = cv.addWeighted(img, 0.7, blank_bg, 0.3, 0)[mask]

cv.imshow('background with watermark',img)
cv.waitKey(0)
cv.destroyAllWindows()
```
